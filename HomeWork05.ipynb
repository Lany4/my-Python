{‘1840e0b9d4’: ‘Продукты’, …}

```python
import json

purchases = {}     # Заведем словарь

f = open('purchase_log.txt', 'r')     # Откроем файл при помощи контекстного менеджера
for line in f:     # Построчно
  line = line.strip()     # Уберем нежелательные пробелы с обех концов строк
  dict_ = json.loads(line)     # конвертируем строку в словарь
  for key, val in  dict_.items():     # Обработаем полученный словарь: значение ключа станет ключем, значение словаря - словарем
    if dict_.get('user_id') != 'user_id' and  dict_.get('category') != 'category':   # Без заголовков
      purchases.setdefault(dict_.get('user_id'), dict_.get('category'))






```

✅ Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки, если покупка была, сам файл visit_log.csv* изменять не надо. Запишите в файл funnel.csv визиты из файла visit_log.csv*, в которых были покупки с указанием категории.
Учтите условия на данные:
содержимое purchase_log.txt* помещается в оперативную память компьютера;
содержимое visit_log.csv* — нет; используйте только построчную обработку этого файла.

```python
i = 0
with open('visit_log.csv', 'r') as f:
  with open('funnel.csv', 'w') as f2:

    for line in f:
      line = line.strip().split(',')
      user_id = str(line[0])
      source = str(line[1])
      if user_id in purchases.keys():
        line_res = user_id + ',' + source + ',' + purchases[user_id] +'\n'
        # i += 1
        # if i > 10:
        #   break
        f2.write(line_res)
```
