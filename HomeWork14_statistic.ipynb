Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
# имеем 300 лошадок
df_horse.shape
```

```python
df_horse.describe()
```

```python
#Изучим процент пропусков
print((df_horse['surgery?'].isna().mean()*100).round(2))       # Не критично
print((df_horse['Age'].isna().mean()*100).round(2))     # Нет пропусков
print((df_horse['rectal temperature'].isna().mean()*100).round(2))      # Критично
print((df_horse['pulse'].isna().mean()*100).round(2))     # Критично
print((df_horse['respiratory rate'].isna().mean()*100).round(2))      # Критично
print((df_horse['temperature of extremities'].isna().mean()*100).round(2))    # Критично
print((df_horse['pain'].isna().mean()*100).round(2))      # Критично
print((df_horse['outcome'].isna().mean()*100).round(2))      # Не критично
```

```python
# Рассмотрим некритичные проценты пропусков
# Для столбца surgery? и outcome по 1 лошадке с пропусками
# Посмотрим эти данные ближе
df_horse[df_horse['surgery?'].isna()]
```

```python
# Так как оба пропуска относятся к одной записи, можем пожертвовать этой записью, так как в дальнейшем отсутствие достоверной информации
# об операции и исходе лошадки нам не позволят сделать корректные выводы и повлияют на статистику
df_horse = df_horse.dropna(subset=['surgery?', 'outcome'])
df_horse.describe()
```

```python
# Замечаем, что есть записи с отсутствием информации в 5 колонках. Для анализа это критично, поэтому удалим такие записи из нашего дф.
df_horse.dropna(thresh = 5, inplace=True)
df_horse.shape          # Удалено 299-289 = 10 записей
```

```python
# посмотрим на весь дф, где значения ректальной температуры nan
tmp = df_horse[df_horse['rectal temperature'].isna()]
tmp.head(10)
```

```python
# Пропуски ректальной температуры мы можем заполнить медианным значением ректальной температуры лошадок по возрастам :)
# А так же ректальная температура может зависеть от того, была ли проведена операция.
df_horse['rectal temperature'].fillna(df_horse.groupby(['surgery?','Age'])['rectal temperature'].transform('median'), inplace=True)
df_horse[df_horse['rectal temperature'].isna()]         # Пропусков больше нет
```

```python
# посмотрим на весь дф, где значения пульса nan
tmp = df_horse[df_horse['pulse'].isna()]
tmp
```

```python
# Пульс отсутствует у лошадок, которым 1 год, заполним его медианным значением пульса с группировкой по опреации, году
df_horse['pulse'].fillna(df_horse.groupby(['surgery?', 'Age'])['pulse'].transform('median'), inplace=True)
df_horse[df_horse['pulse'].isna()]     # Пропусков больше нет
```

```python
# посмотрим пропуски по частоте дыхания
tmp = df_horse[df_horse['respiratory rate'].isna()]
tmp.head(5)
```

```python
# Частота дыхания однозначно связана с пульсом.
# разделим значения пульса на 4 равные группы по кол-ву значений
groups_label =['1', '2', '3', '4']
pulse_groups = pd.qcut(df_horse['pulse'], 4, labels = groups_label)

# pulse_groups    [(29.999, 48.0] < (48.0, 64.0] < (64.0, 88.0] < (88.0, 146.0]]
pulse_groups
```

```python
# посчитаем медианные значения частоты дыхания по группам пульса
# добавим временный столбец с номером группы пульса
df_horse['pulse_group'] = pulse_groups
df_horse.head(5)
```

```python
# заполним пропуски частоты дыхания медианными значениями частоты дыхания по группам пульса
df_horse['respiratory rate'].fillna(df_horse.groupby('pulse_group')['respiratory rate'].transform('median'), inplace=True)
df_horse[df_horse['respiratory rate'].isna()]     # Пропусков больше нет
# Округлим значения частоты дыхания до 1 pyfrf
df_horse['respiratory rate'] = df_horse['respiratory rate'].round(1)
# Удалим временный столбец
df_horse.drop('pulse_group', axis=1, inplace=True)
df_horse.head(5)
```

```python
# посмотрим на весь дф, где значения температуры конечностей nan
tmp = df_horse[df_horse['temperature of extremities'].isna()]
tmp.head(5)
# по условию для колонки : горячие конечности должны коррелировать с повышенной ректальной температурой. (норм 37,8)
```

```python
# сначала заполним пропуски в тех местах, где ректальная температура повышена (2)
df_horse.loc[df_horse['rectal temperature']>37.8, 'temperature of extremities'] = df_horse.loc[df_horse
 ['rectal temperature']>37.8, 'temperature of extremities'].fillna(2)
```

```python
# остались лошадки с нормальной ректальной температурой - 16 штук
# будем считать, что при нормальной ректальной температуре и конечности нормальной температуры (1)
df_horse['temperature of extremities'].fillna(1, inplace=True)
df_horse[df_horse['temperature of extremities'].isna()]       # Пропусков больше нет

```

```python
# посмотрим на весь дф, где значения боли nan
df_horse[df_horse['pain'].isna()]
# по описанию, если операции не было, показатель должен быть 4-5
df_horse.loc[df_horse['surgery?'] == 1, 'pain'] = df_horse.loc[df_horse['surgery?'] == 1, 'pain'].fillna(4)
# предшествующее лечение может маскировать боль, значит  1-2
df_horse.loc[df_horse['surgery?'] == 2, 'pain'] = df_horse.loc[df_horse['surgery?'] == 2, 'pain'].fillna(4)
df_horse[df_horse['pain'].isna()]     # Пропусков больше нет
```
