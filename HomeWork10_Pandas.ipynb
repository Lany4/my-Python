**Задание 1**

Скачайте с сайта датасет любого размера. Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
movies = pd.read_csv('movies.csv')
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
# 1й вариант
df = movies.merge(ratings, on='movieId', how='left')
a = df.groupby(['title', 'rating']).count()
a[a['movieId'] == a['movieId'].max()]
```

```python
# 2й вариант (а вот как сделать лаконично и коротко, поже, не сообразила)

df = movies.merge(ratings, on='movieId', how='left')
df.filter(items = ['title', 'rating']).query('rating == 5.0').groupby('title').count().sort_values(by='rating', ascending = False).head(1)
```

```python
#От преподавателя
df[df.rating == 5].movieId.value_counts()
```
