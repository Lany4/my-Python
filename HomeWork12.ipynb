✅ **Задание 1**

Для датафрейма log из материалов занятия создайте столбец source_type по правилам:

если источник traffic_source равен Yandex или Google, то в source_type ставится organic;

для источников paid и email из России ставим ad;

для источников paid и email не из России ставим other;

все остальные варианты берём из traffic_source без изменений.

```python
df = pd.read_csv('visit_log.csv', sep = ';')
df.head(2)
```

```python
df.loc[(df['traffic_source'].str.contains('Yandex|Google', case = False)), 'source_type'] = 'organic'
df.loc[((df['traffic_source'].str.contains('paid|email', case = False) & df['region'].str.contains('Russia', case = False))
        ), 'source_type'] = 'ad'
df.loc[((df['traffic_source'].str.contains('paid|email', case = False) & ~(df['region'].str.contains('Russia', case = False)))
        ), 'source_type'] = 'other'
df['source_type'].fillna(df['traffic_source'], inplace=True)
```

✅ **Задание 3**

Используйте файл с оценками фильмов ml-latest-small/ratings.csv.

Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. *Под временем жизни понимается разница между максимальным и минимальным значениями столбца timestamp для данного значения userId*.

```python
df = pd.read_csv('ratings.csv')
df.head()
```

```python
tmp = df.groupby('userId').count().reset_index()

mask = tmp['movieId'] > 100    # сделала маску для тех пользователей,кто поставил больше 100 оценок
mask.head()
```

```python
df1 = df.groupby('userId')[['movieId', 'timestamp']].apply(lambda x: x.max() - x.min()).reset_index()     # сгруппировала по пользовтелю и применила функцию размах
rez = df1[mask][['userId', 'timestamp']].rename(columns={'timestamp': 'avg_lifetime'})  # к сгруппированному ДФ применила маску(тех, кто>100) и переименовала колонку
rez.sort_values(by='avg_lifetime', ascending=False)
```

✅ **Задание 4**

Дана статистика услуг перевозок клиентов компании по типам (см. файл “Python_13_join.ipynb” в разделе «Материалы для лекции “Продвинутый pandas”» ---- Ноутбуки к лекции «Продвинутый pandas»).

Нужно сформировать две таблицы:

таблицу с тремя типами выручки для каждого client_id без указания адреса клиента;

аналогичную таблицу по типам выручки с указанием адреса клиента.

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
```

```python
table1 = client_base.merge(rzd, how='left').merge(auto, how='left').merge(air, how='left')[['client_id', 'rzd_revenue', 'auto_revenue', 'air_revenue']].fillna(0)
table1
```

```python
table2 = client_base.merge(rzd, how='left').merge(auto, how='left').merge(air, how='left')[['client_id', 'address', 'rzd_revenue', 'auto_revenue', 'air_revenue']].fillna(0)
table2
```

**добавила в таблицу суммы по строкам и столбцам**

```python
table2['sum_revenue'] = table2[['rzd_revenue','auto_revenue', 'air_revenue']].sum( axis=1)     # сумма по каждому клиенту
# дф из словаря с подсчетом по колонкам
rez =  pd.DataFrame.from_dict(
    {'client_id': ['-'], 'address': ['-'], 'rzd_revenue': [table2['rzd_revenue'].sum()], 'auto_revenue': [table2['auto_revenue'] .sum()],
     'air_revenue': [table2['air_revenue'].sum()], 'sum_revenue': [table2['sum_revenue'].sum()]}
)
# объединение дф со словарём
table = pd.concat([table2, rez], ignore_index=True)
table
```
