✅ Проверить гипоезу о том, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом

* H0: В среднем работники с любым окладом проводят на работе одинаковое количество времени
* H1: В среднем работники с разными окладами проводят на работе разное количество времени

α = 0.05

```python
df_high_salary = df[df['salary'] == 'high']
df_low_salary = df[df['salary'] == 'low']
plt.hist(df_low_salary['average_montly_hours'])
stat, p = st.shapiro(df_high_salary['average_montly_hours'])
if p > 0.05:
  print(f'Данные, вероятно, распределены нормально, pval = {p:.3f}')
else:
  print(f'Данные, вероятно, распределены не нормально, pval = {p:.3f}')
```

```python
from scipy.stats import mannwhitneyu
stat, p = mannwhitneyu(df_high_salary['average_montly_hours'],
                       df_low_salary['average_montly_hours'])
if p > 0.05:
  print(f'Принимаем нулевую гипотезу, средние, вероятно, статистически не различаются. pval = {p:.3f}')
else:
  print(f'Отвергаем нулевую  в пользу альтернативной, т.к. средние, вероятно, статистически различны. pval = {p:.3f}')

tmp = pd.concat([df_high_salary['average_montly_hours'],
                 df_low_salary['average_montly_hours']],
                 axis = 1)
tmp.columns = ['avg_mon_hours_high_salary', 'avg_mon_hours_low_salary']
sns.boxplot( data=tmp )
plt.show()
a = pd.concat([tmp['avg_mon_hours_high_salary'].describe(),
               tmp['avg_mon_hours_low_salary'].describe()],
               axis = 1)
a
```

**Гипоезу о том, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом опровергаем в пользу гипотезы о том, что сотрудники с высоким окладом статистически проводят столько же времени на работе, сколько и сотрудники с низким окладом.**

✅ Рассчитать следующие показатели среди уволившихся и не уволившихся сотрудников (по отдельности):
* Доля сотрудников с повышением за последние 5 лет
* Средняя степень удовлетворенности
* Среднее количество проектов

```python
left_empl = df[df['left'] == 1]
not_left_empl = df[df['left'] == 0]
```

```python
result = pd.DataFrame(columns = ['Показатель', 'Уволившиеся', 'Не уволившиеся'])
result.loc[len(result.index)] = ['Доля сотрудников с повышением за последние 5 лет',
                                 (left_empl[left_empl['promotion_last_5years'] == 1]['left'].count() * 100
                                        / left_empl['left'].count()).round(2),
                                 (not_left_empl[not_left_empl['promotion_last_5years'] == 1]['left'].count() * 100
                                        / not_left_empl['left'].count()).round(2)]
result.loc[len(result.index)] = ['Средняя степень удовлетворенности',
                                 (left_empl['satisfaction_level'].mean()).round(2),
                                 (not_left_empl['satisfaction_level'].mean()).round(2)]
result.loc[len(result.index)] = ['Среднее количество проектов',
                                 (left_empl['number_project'].mean()).round(2),
                                 (not_left_empl['number_project'].mean()).round(2)]
result
```

✅ 1. Разделить данные на тестовую и обучающую выборки


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df[['satisfaction_level', 'last_evaluation', 'number_project',
                'average_montly_hours', 'time_spend_company', 'Work_accident', 'promotion_last_5years']]
y = df['left']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

✅ 2. Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary)


```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
y_train_pred = lm.predict(X_train)
y_test_pred = lm.predict(X_test)
```

✅ 3. Оценить качество модели на тестовой выборке

```python
import sklearn.metrics as metrics
print(f'R2_test: {metrics.r2_score(y_test, y_test_pred):.3f}')
```

```python
sns.histplot((y_test-y_test_pred), kde=True, bins=20)
```

```python
pd.DataFrame(lm.coef_ , X.columns, columns=['Коэффициенты'])
```

Низкий показатель коэффициента детерминации (0.183) в данном случае может означать, что данные не могут быть достаточно хорошо смоделированы на основе линейной зависимости
