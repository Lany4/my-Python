✅**Задание 1**

Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:

оценка 2 и ниже — низкий рейтинг;
оценка 4 и ниже — средний рейтинг;
оценка 4.5 и 5 — высокий рейтинг.

Результат классификации запишите в столбец class.

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
df = movies.merge(ratings, on='movieId', how='left')
df.head()
```

```python
def rating(num):
  """ Функция приничает число (рейтинг) и возвращает уроверь рейтинга: высокий, средний, низкий """
  if num <= 2:
    return 'low'
  elif num <= 4:
    return 'medium'
  elif num == 4.5 or num == 5:
    return 'high'
```

```python
df['class'] = df['rating'].apply(rating)
df.head()
```

✅**Задание 2**

Используйте файл keywords.csv.

Нужно написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определённому региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

1 'Центр': ['москва', 'тула', 'ярославль'],

2'Юг': ['краснодар', 'ростов', 'волгоград']

3'Северо-Запад': ['петербург', 'псков', 'мурманск'],

4'Сибирь': ['иркутск', 'осмк', 'томск']

5'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

```python
geo = pd.read_csv('keywords.csv')
geo.head()
```

```python
# добавила 2й и пятый регион...
def classifyer(ask):
  """ функция принимает на вход строку и по наличию в ней города, возвращает название региона"""
  ask = ask.strip().split(' ')
  geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
              'Юг': ['краснодар', 'ростов', 'волгоград'],
              'Северо-Запад': ['петербург', 'псков', 'мурманск'],
              'Сибирь': ['иркутск', 'осмк', 'томск'],
              'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
  flag = False
  for el in ask:
    for k, v in geo_data.items():
      if el in v:
        flag = True
        key = k
  if flag == True:
    return key
  else:
    return 'undefined'
```

```python
# применим классификацию к стобцу keyword

geo['region'] = geo['keyword'].apply(classifyer)
geo.head()
```

```python
# проверим результат
geo.groupby('region').count()
```

✅**Задание 3 (бонусное)**

Есть мнение, что раньше снимали настоящее кино, не то что сейчас. Ваша задача — проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия: файл movies.csv и ratings.csv из базы. Нужно проверить, верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

Вы не будете затрагивать субьективные факторы выставления этих рейтингов, а пройдётесь по алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010 года.

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте алгоритм:

  1)для каждой строки пройдите по всем годам списка years

  2)если номер года присутствует в названии фильма, то функция возвращает этот год, как год выпуска;

  3)если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год.

3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’.

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга.

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
df = movies.merge(ratings, on='movieId', how='left')
df.head()
```

```python
def year_catch(title):
  years = [str(i) for i in range(1950, 2011)]
  year_find = re.findall(r'[0-9]{4}', title)
  year_find = ''.join(year_find)
  if year_find in years:
    return year_find
  else:
    return '1900'
```

```python
df['year'] = df['title'].apply(year_catch)
df.head()
```

```python
#для начала отсортировала по увеличению года и построила график.
# рейтинг с годами падает, похоже, гипотеза подтвердилась
df[['year', 'rating']].groupby('year').mean('rating').sort_values(by='year')
```

```python
#сортировка по убыванию среднего рейтинга
df[['year', 'rating']].groupby('year').mean('rating').sort_values(by='rating', ascending = False)
```
